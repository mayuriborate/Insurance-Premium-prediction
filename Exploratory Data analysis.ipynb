```python
### Now we draw the relational plot
### in insurance data the target variable is 'expenses' so we check the relation between target and all other features 
       ###step by step
```

```python
### first relation between categorical and numerical
```

```python
sns.swarmplot('sex','expenses',data=idata)
plt.title('expenses and sex')
plt.show()
```

```python
sns.swarmplot('smoker','expenses',data=idata)
plt.title('expenses and smoker')
plt.show()
```

```python
sns.swarmplot('children','expenses',data=idata)
plt.title('expenses and children')
plt.show()
```

```python
idata.columns
```

```python
sns.swarmplot('region','expenses',data=idata)
plt.title('expenses and region')
plt.show()
```

```python
#### relation between numerical and numerical
```

```python
sns.scatterplot('age','expenses',data=idata)
plt.title('expenses and age')
plt.show()
```

```python
sns.scatterplot('bmi','expenses',data=idata)
plt.title('expenses and bmi')
plt.show()
```
